```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
#PLEASE CHECK LOCATION BEFORE RUNNING
df = pd.read_csv("data.csv")
df.head()
```

```python
#Shape of the dataset
df.describe()
```

```python
ROA = df[[" ROA(C) before interest and depreciation before interest",
          " ROA(A) before interest and % after tax",
          " ROA(B) before interest and depreciation after tax"]]

corr=ROA.corr()
sns.heatmap(corr, vmin=-1,vmax=1)
```

```python
fig, ax = plt.subplots(figsize=(20, 12))
# plot heatmap
sns.scatterplot(data=df, y=" Total Asset Growth Rate", x=" ROA(A) before interest and % after tax", hue="Bankrupt?",style="Bankrupt?",
                size="Bankrupt?", sizes=(150,5))
plt.axvline(0.55, color='r')
plt.title ("Total Asset Growth Rate v/s ROA(A)")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 12))
# plot heatmap
sns.scatterplot(data=df, y=" Total Asset Growth Rate", x=" ROA(C) before interest and depreciation before interest", hue="Bankrupt?",
                style="Bankrupt?",
                size="Bankrupt?", sizes=(150,5))
plt.axvline(0.54, color='r')
plt.title ("Total Asset Growth Rate v/s ROA(C)")
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(20, 12))
# plot heatmap
sns.scatterplot(data=df, y=" Total Asset Growth Rate", x=" ROA(B) before interest and depreciation after tax", hue="Bankrupt?",style="Bankrupt?",
                size="Bankrupt?", sizes=(150,10))
plt.axvline(0.55, color='r')
plt.title ("Total Asset Growth Rate v/s ROA(B)")
plt.show()
```

```python
bkr= df["Bankrupt?"]==1
mean_for_interest_bkr = df.loc[bkr, ' Interest-bearing debt interest rate'].mean()

not_bkr= df["Bankrupt?"]==0
mean_for_interest_not_bkr = df.loc[not_bkr, ' Interest-bearing debt interest rate'].mean()

print("Mean Interest for Bankrupt Companies : ",mean_for_interest_bkr,"\nMean Interest for Non- Bankrupt Companies: ",mean_for_interest_not_bkr)
```

```python
bkr= df["Bankrupt?"]==1
mean_for_tax_bkr = df.loc[bkr, ' Tax rate (A)'].mean()

not_bkr= df["Bankrupt?"]==0
mean_for_tax_not_bkr = df.loc[not_bkr, ' Tax rate (A)'].mean()

print("Mean Tax for Bankrupt Companies : ", round(mean_for_tax_bkr*100,2),
      "%\nMean Tax for Non- Bankrupt Companies: ",round(mean_for_tax_not_bkr*100,2),"%")
```
